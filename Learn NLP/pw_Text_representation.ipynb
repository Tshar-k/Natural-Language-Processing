## Word2Vec


```python
import pandas as pd

messages = pd.read_csv("/content/SMSSpamCollection.txt",sep= '\t',names = ['label','message'])
```

```python
messages.head()
```

```python
messages.shape
```

```python
messages['message'].loc[100]
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0,len(messages)):
  review = re.sub('[^a-zA-Z]',' ',messages['message'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 2500,binary = True,ngram_range = (2,2))    # top 2500 features can be used
x = cv.fit_transform(corpus).toarray()
```

```python
x.shape
```

```python
y = pd.get_dummies(messages['label'])
m = y
y = y.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.1,random_state = 42)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detection_model = MultinomialNB().fit(x_train,y_train)
```

```python
y_pred = spam_detection_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_test,y_pred)
```

```python
score
```

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')  # Download the punkt tokenizer

# Sample text data
corpus = "weather is chill and cold by the way and here in chennai it is hot and humid and i dont like the atmosphere and climate in bangalore"

# Tokenize the text into words
tokenized_text = word_tokenize(corpus.lower())  # Convert to lowercase for consistency

# Define Word2Vec model
model = Word2Vec(sentences=[tokenized_text], vector_size=100, window=5, min_count=1, workers=4)

# Training the Word2Vec model
model.train([tokenized_text], total_examples=1, epochs=10)

# Finding the vector for a specific word
word_vector = model.wv['climate']
print("Vector representation of 'word':\n", word_vector)

# Finding similar words
similar_words = model.wv.most_similar('cold', topn=5)
print("\nWords similar to 'word':\n", similar_words)
```
