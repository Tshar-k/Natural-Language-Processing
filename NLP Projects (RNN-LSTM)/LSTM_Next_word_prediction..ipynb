## Tokenizer the data

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
tokenizer.word_index
```

```python
for sentences in faqs.split('\n'):
  print(sentences)
```

```python
for sentences in faqs.split('\n'):
  print(tokenizer.texts_to_sequences([sentences])[0])
```

```python
input_sequences = []
for sentences in faqs.split('\n'):
  tokenized_sentences = tokenizer.texts_to_sequences([sentences])[0]
  for i in range(1,len(tokenized_sentences)):
    n_gram = tokenized_sentences[:i+1]
    input_sequences.append(n_gram)
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
```

## input output segmentations

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences,maxlen = max_len,padding = 'pre')
```

```python
x = padded_input_sequences[:,:-1]
```
